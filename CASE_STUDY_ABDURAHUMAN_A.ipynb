```python
import numpy as np
```

```python
house['TotalCharges'] = pd.to_numeric(house['TotalCharges'], errors='coerce')
house.fillna({'TotalCharges': house['TotalCharges'].median()}, inplace=True)

q1 = np.quantile(house['TotalCharges'],0.25)
q3 = np.quantile(house['TotalCharges'],0.75)

iqr = q3 - q1
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)

print(f"Upper Bound: {upper_bound}")
print(f"Lower Bound: {lower_bound}")
print(f"IQR: {iqr}")
```

```python
tot_char_out = []
for i in house['TotalCharges']:
  if i > upper_bound or i < lower_bound:
    tot_char_out.append(i)
```

```python
house['TotalCharges']=house['TotalCharges'].clip(lower=lower_bound,upper=upper_bound)
```
