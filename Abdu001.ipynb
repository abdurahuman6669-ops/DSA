```python
median_age = df['age'].median()
df['age'].fillna(median_age, inplace=True)
```

8

```python
def get_age_group(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(get_age_group)
df['age_group'] = np.where(df['age'] < 18, 'Child',
                    np.where(df['age'] <= 60, 'Adult', 'Senior'))

```
